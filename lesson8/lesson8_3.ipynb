```python
def menu(wine, entree='牛排', dessert='蛋糕'):
    print(f"酒類:{wine}")
    print(f"主餐:{entree}")
    print(f"點心:{dessert}")
```

```python
menu('白酒')
print("========")
menu('白酒',dessert='冰淇淋')
print("========")
menu('白酒',dessert='冰淇淋',entree='雞排')
```

```python
#python3.8以後才有的
#/前面的定要用引數值的呼叫
def menu(wine, entree='牛排', dessert='蛋糕', /):
    print(f"酒類:{wine}")
    print(f"主餐:{entree}")
    print(f"點心:{dessert}")
```

```python
menu('白酒','雞排','冰淇淋')
print("========")
menu('白酒',dessert='冰淇淋')
```

```python
def menu(wine, entree='牛排',/,dessert='蛋糕'):
    print(f"酒類:{wine}")
    print(f"主餐:{entree}")
    print(f"點心:{dessert}")
```
