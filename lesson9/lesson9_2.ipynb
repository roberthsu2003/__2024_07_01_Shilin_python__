```python
#透過自定的__init__(self),建立自訂的初始化
class Person():
    def __init__(self,n:str):
        self.name = n

    def __repr__(self):
        return f"我是person的實體,我的名字是:{self.name}"

    

p1 = Person("Robert Hsu")
print(p1)


p2 = Person(n="徐國堂")
print(p2)
```

```python
class Student(Person):
    def __init__(self,n:str, ch:int, en:int, ma:int):
        super().__init__(n=n)
        self.chinese = ch
        self.english = en
        self.math = ma

    def __repr__(self):
        return "我是Student的實體"

s1 = Student("張同學",ch=78,en=92,ma=69)
print(s1.name)
print(s1.chinese)
print(s1.english)
print(s1.math)
print(s1)

print("=================")
s2 = Student("王同學",ch=92,en=79,ma=88)
print(s2.name)
print(s2.chinese)
print(s2.english)
print(s2.math)
print(s2)
```
