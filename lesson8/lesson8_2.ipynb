```python
def menu(wine, entree, dessert):
    print(f"酒類:{wine}")
    print(f"主餐:{entree}")
    print(f"點心:{dessert}")
```

```python
#引數值的呼叫,要依照參數的順序
menu('白酒','牛排',"蛋糕")
```

```python
#引數名稱的呼叫,可以不依照參數的順序
menu(entree='牛排',dessert='蛋糕',wine='白酒')
```

```python
#混合呼叫
#前面一定要引數值呼叫
#後面要引數名稱的呼叫
#使用引數名稱呼叫後,後面全部要使用引數名稱的呼叫

menu('白酒','牛排',dessert='蛋糕')
```

```python
menu('白酒',dessert='蛋糕',entree='牛排')
```
