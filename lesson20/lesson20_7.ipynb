```python
from tools import get_path,station_info,merge_dataFrame
import pandas as pd

fileName_list = get_path()
stations_df = station_info()
all_in_out = merge_dataFrame(fileName_list, stations_df)
station_in_out= pd.concat(all_in_out)
station_in_out1 = station_in_out.sort_values(by='日期')
station_in_out1
```

```python
def top(df, n=5, column='進站'):
    return df.sort_values(by=column,ascending=False)[:n]

top(station_in_out1)
```

```python
years_months = station_in_out1['日期'].dt.strftime('%Y-%m')
station_in_out2 = station_in_out1.groupby(by=[years_months,'站名'])[['進站','出站']].apply(top)
station_in_out2
```

```python
station_in_out2 = station_in_out1.groupby(by=[years_months,'站名'])[['進站','出站']].apply(top,n=1)
station_in_out2
```

```python
station_in_out2 = station_in_out1.groupby(by=[years_months,'站名'])[['進站','出站']].apply(top,n=1,column='出站')
station_in_out2
```
